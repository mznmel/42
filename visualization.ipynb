# Visualization

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import product
import math
```

```python
x = np.linspace(0, 360, 100)
```

```python
def plot_cos_func():
    for i in range(3):
        plt.plot(x, np.cos((x*(i+1))*np.pi/180))
```

```python
plt.figure(figsize=(16, 5))
plot_cos_func()
```

```python
styles = ["white", "dark", "whitegrid", "darkgrid", "ticks"]
contexts = ["notebook"] #["paper", "notebook", "talk", "poster"]
call_despine = [True, False]

all_combinations = list(product(styles, contexts, call_despine))
columns = 2
rows = math.ceil(len(all_combinations) / 2)

sns.set(rc={"figure.figsize": (16*columns, 5*rows)})

for i, (style, context, despine) in enumerate(all_combinations):
    plt.subplot(rows, columns, i+1)
    sns.set_style(style)
    sns.set_context(context)
    plot_cos_func()
    if despine:
        sns.despine()
    plt.title("Style: {}, Context: {}, Despine: {}".format(style, context, despine))
```
