# Visualization

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 360, 100)
```

```python
def plot_cos_func():
    plt.figure(figsize=(16, 5))
    for i in range(3):
        plt.plot(x, np.cos((x*(i+1))*np.pi/180))
```

```python
plot_cos_func()
```

```python
for style in ["white", "dark", "whitegrid", "darkgrid", "ticks"]:
    for despine in [True, False]:
        sns.set_style(style)
        plot_cos_func()
        if despine:
            sns.despine()
        plt.title("Style:{}, Despine:{}".format(style, despine))
```
